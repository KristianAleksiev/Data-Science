```python
%matplotlib inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

### Styling the plots

```python
plt.style.use("seaborn")
```

## Data visualization, EDA

```python
data = np.random.randint(1, 11, 50)
```

```python
data
```

### Plotting a histogram

```python
plt.hist(data)
plt.show()
```

### Plotting a boxplot

```python
plt.boxplot(data)
plt.show()
```

### Plotting a bar chart

```python
plt.bar(data, height=200)
plt.show()
```

```python
plt.barh(data, width=200)
plt.show()
```

### Plotting a pie chart

```python
plt.pie(data)
plt.show()
```
